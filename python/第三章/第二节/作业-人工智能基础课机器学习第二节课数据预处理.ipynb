# 1. 从fundemantals.csv开始！

fundemantals.csv 是这些股票的年报数据

请用数据回答以下问题：

1. S&P500股票在2015年`net income`的均值是多少？最大值比最小值多多少？（每问10分，共计20分）
2. S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是多少？固定资产占总资产比例最小的股票是的代码（ticker symbol）是什么？（每问10分，共计20分）


```python
#1.S&P500股票在2015年net income的均值是多少？最大值比最小值多多少？
fundamentals_df= pd.read_csv('./nyse/fundamentals.csv')
#print(fundamentals_df.shape)
fundamentals_df['Period Ending'] = pd.to_datetime(fundamentals_df['Period Ending'])
fundamentals_df = fundamentals_df.set_index('Period Ending')
#print(fundamentals_df.index)
#print(type(fundamentals_df.index))
#fundamentals_df['2015']['Net Income'].head()
net_income_2015 = pd.DataFrame(fundamentals_df['2015']['Net Income'])
#net_income_2015
#print(net_income_2015.shape)
#net_income.head()
func_range = lambda x: x.max() - x.min()
func_net_income_2015 = lambda x: pd.Series([x.min(), x.max(),func_range(x),x.mean()], index=['min', 'max', 'range','mean'])
net_income_2015_df=net_income_2015.apply(func_net_income_2015)
print('S&P500股票在2015年net income的均值:',net_income_2015_df.loc['mean','Net Income'])
print('最大值比最小值:',net_income_2015_df.loc['range','Net Income'])
net_income_2015_df
```

```python
#2.S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是多少？固定资产占总资产比例最小的股票是的代码（ticker symbol）是什么？
assets_2016 = pd.DataFrame(fundamentals_df['2016'][['Ticker Symbol','Fixed Assets','Total Assets']])
assets_2016['fixed_total_ratio'] = assets_2016.apply(lambda x: x[1]/x[2], axis=1)
print('在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值:',assets_2016['fixed_total_ratio'].mean())
print('固定资产占总资产比例最小的股票是的代码:',assets_2016['Ticker Symbol'][assets_2016['fixed_total_ratio']==assets_2016['fixed_total_ratio'].min()].values[0])
assets_2016
```

# 2. 加入securities.csv~

securities.csv包含了这些股票的基本信息

1. 请列举出各个sector中的加入时间最早的股票名称（10分）
2. 请列举出每一个州中加入时间最晚的股票名称（10分）

```python
securities_df= pd.read_csv('./nyse/securities.csv')
#securities_df.head()
securities_df=securities_df.dropna()
#securities_df.head()
securities_df['Date first added'] = pd.to_datetime(securities_df['Date first added'])

securities_df_earlist_sector=securities_df.groupby(securities_df['GICS Sector']).apply(lambda x: x[x['Date first added']==x['Date first added'].min()])
securities_df_earlist_sector.sort_values(by=['Date first added'], ascending=True, inplace=True)
#print(securities_df_earlist_sector.loc[:,['GICS Sector','Ticker symbol']])
print('{0}各个sector中的加入时间最早的股票名称{0}\n'.format('*'*30))
securities_df_earlist_sector[['Ticker symbol','Date first added']]
```

```python
def getState(row):
    state=row[5].split(',')[1]
    if state.find('[')!=-1:
         return state[0:state.index('[')]
    else:
        return state
securities_df['State'] = securities_df.apply(getState, axis=1)
#securities_df
securities_df_earlist_state=securities_df.groupby(securities_df['State']).apply(lambda x: x[x['Date first added']==x['Date first added'].max()])
#securities_df_earlist_state
securities_df_earlist_state.sort_values(by=['Date first added'], ascending=False, inplace=True)
print('{0}每一个州中加入时间最晚的股票名称{0}\n'.format('*'*30))
securities_df_earlist_state[['Ticker symbol','Date first added']]
```

# 3. merge!

现在你需要同时处理来自两个表中的信息了

1. 请思考，合并两个表的信息的时候，我们应该用什么样的准则对其它们（10分）

   答：merge是按照两个dataframe共有的column进行连接，两个dataframe必须具有同名的column
       how=inner,结果中的键是df1和df2的交集
       how=outter,结果中的键是df1和df2的并集
       how=left，就是让df1保留所有的行列数据，df2根据df1的行列进行补全。
       how=right，就是让df2保留所有的行列数据，df1根据df2的行列进行补全。
       on参数可以指定merge的列名
       
   
2. 请列举每个sector在2013-2016年累计Research&Development的总投入（10分）
3. 请列举出每个sector中，在2013-2016年累计Research&development投入最大的3家公司的名称以及投入的数值（20分）


```python
fundamentals_securities_df=pd.merge(pd.read_csv('./nyse/fundamentals.csv')[['Ticker Symbol','Period Ending','Research and Development']], 
                                    securities_df[['Ticker symbol','Security','GICS Sector']].rename(columns={'Ticker symbol':'Ticker Symbol'}), 
                                    how='left',
                                    on='Ticker Symbol')
fundamentals_securities_df['Period Ending'] = pd.to_datetime(fundamentals_securities_df['Period Ending'])
fundamentals_securities_df = fundamentals_securities_df.set_index('Period Ending')
fundamentals_securities_df = fundamentals_securities_df['2013':'2016'].dropna()

print('{0}每个sector在2013-2016年累计Research&Development的总投入{0}\n'.format('*'*30))
pd.DataFrame(fundamentals_securities_df['Research and Development'].groupby(fundamentals_securities_df['GICS Sector']).sum())
```

```python
securities_sector_df=pd.merge(fundamentals_securities_df['Research and Development'].groupby(fundamentals_securities_df['Security']).sum(),
                              securities_df[['Security','GICS Sector']].drop_duplicates(),
                              how='left',
                              on='Security')
#securities_sector_df
def top(df, n=3, column='Research and Development'):
    return df.sort_values(by=column)[-n:]
print('{0}每个sector中，在2013-2016年累计Research&development投入最大的3家公司的名称以及投入的数值{0}\n'.format('*'*25))
securities_sector_df.groupby(securities_sector_df['GICS Sector']).apply(top)
```

# 4. 现在让我们来看看更加复杂的数据

请导入price.csv，然后结合你的聪明才智回答以下问题（附加题，40分）

假设你是某基金公司的老板，现在对于每只股票，你都专门安排了一位负责它的交易员。公司规定每一位交易员手中的资金要么全部买入要么全部卖出（空仓，转化为现金）。假设2016年每一位交易员手中都有10000美元，假设他们都能够看到2016年全年的数据，假设他们都能抓住每一次机会，那么请问2016年底时，赚钱最多的股票是哪一只，赚了多少钱？

```python
prices_df=pd.read_csv('./nyse/prices.csv')
prices_df['date'] = pd.to_datetime(prices_df['date'])
prices_df = prices_df.set_index('date')
prices_df = prices_df['2016'].dropna()
base=10000
def getFinalTurnover(symbol):
    symbol_df=prices_df.loc[prices_df['symbol'] == symbol]
    symbol_df['turnover']=0
    for i in range(0, len(symbol_df)):
        symbol_df['turnover']=(symbol_df['turnover'].shift(1).fillna(base)/symbol_df['low'])*symbol_df['high']
    return symbol,symbol_df.iloc[len(symbol_df)-1].at['turnover']-base

symbol_list=prices_df['symbol'].drop_duplicates().tolist()
symbol_final_turnover=list(map(getFinalTurnover,symbol_list))
#print(symbol_final_turnover)
symbol_final_turnover_sort= sorted(symbol_final_turnover, key=lambda d:d[1], reverse = True)
#print(symbol_final_turnover_sort)
print('2016年底时，赚钱最多的股票是{0}，赚了{1}'.format(symbol_final_turnover_sort[0][0],symbol_final_turnover_sort[0][1]))
symbol_final_turnover_df=pd.DataFrame(symbol_final_turnover_sort,columns=['symbol','final_turnover'])
symbol_final_turnover_df

#getFinalTurnover('WLTW')
```
