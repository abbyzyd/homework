## 1. 实现一周的记账功能

首先记录输入的7天的收入，然后记录输入的7天的支出。  
然后打印出以下内容：
- 7天的收入(list)
- 7天的支出(list)
- 每天的结余(dict)
- 最终的结余

````python
list_days=['星期天','星期一','星期二','星期三','星期四','星期五','星期六']
list_income=[]
list_outcome=[]
dict_daily_balance={}
total_income=0
total_outcome=0
total_balance=0
for i,val in enumerate(list_days):
    income=input('请输入{0}的收入(为空默认为0.00)：'.format(list_days[i]))
    if income==None or income=='':
        income=0
    else:
        income=float(income)
    outcome=input('请输入{0}的支出(为空默认为0.00)：'.format(list_days[i]))
    if outcome==None or outcome=='':
        outcome=0
    else:
        outcome=float(outcome)
    total_income+=income
    list_income.append(income)
    total_outcome+=outcome
    list_outcome.append(outcome)
    dict_daily_balance[list_days[i]]=income-outcome
    total_balance+=income-outcome

print('7天的收入:',list_income)

print('7天的支出:',list_outcome)

print('每天的结余',dict_daily_balance)

print('最终的节约',total_balance)

    
````

## 2. 实现ATM机器
- 查询余额
- 存款
- 取款
- 退出

````python
dict_function={'1':'查询余额','2':'存款','3':'取款','4':'退出'}
balance=100
str_function=str(dict_function)
function_no=input('请选择功能{0}：'.format(str_function))
while True:
    
    if function_no not in dict_function.keys():
        function_no=input('输入有误，请重新选择：')
        continue
    else:
        if function_no=='1':
            print('您当前的余额：{0}'.format(balance))
            function_no=input('请继续选择需要的功能{0}：'.format(str_function))
        elif function_no=='2':
            money=float(input('请输入{0}金额：'.format(dict_function[function_no])))
            balance+=money
            print('已成功{0}{1}'.format(dict_function[function_no],money))
            function_no=input('请继续选择需要的功能{0}：'.format(str_function))
        elif function_no=='3':
            money=float(input('请输入{0}金额：'.format(dict_function[function_no])))
            balance-=money
            print('已成功{0}{1}'.format(dict_function[function_no],money))
            function_no=input('请继续选择需要的功能{0}：'.format(str_function))
        else:
            print('感谢您的支持，再见！')
            break
        
````
